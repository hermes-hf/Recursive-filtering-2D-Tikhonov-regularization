# Multigrid Solver

```julia
function cm(x,n)
    return mod(x-1,n) +1
end

#values lx ly to use according to image scale (avoid excessive smoothing)
function getTikhonovMatrix(img,lx,ly; homog=1)
    n,m = size(img)
    A = zeros(n*m,n*m)
    
    Wx = zeros(n,m)
    Wy = zeros(n,m)
    
    function imf(x,y)
        if x>n || x<1 || y>m || y<1
            return 0 
        end
        return img[x,y]
    end
    
    for x=1:n, y=1:m
        Wx[x,y] = lx/(eps()+abs(imf(x,y)-imf(x-1,y)))
        Wy[x,y] = ly/(eps()+abs(imf(x,y)-imf(x,y-1)))
        if homog==1
            Wx[x,y] = lx
            Wy[x,y] = ly
        end
    end
    
    function wx(x,y)
        if x>n || x<1 || y>m || y<1
            return 0 
        end
        return Wx[x,y]
    end
    function wy(x,y)
        if x>n || x<1 || y>m || y<1
            return 0 
        end
        return Wy[x,y]
    end
    
    function st(A,val,i,j,x,y)
        if i>n*m || i<1 || j> n*m || j<1
            return 
        end
        if x>n || x<1 || y>m || y<1
            return 
        end
        A[i,j] = val
        return
    end
    
    #fill matrix
    for x = 1:n, y=1:m
        i = (y-1)*n + x
        A[i,i] = 1 + wx(x,y) + wx(x+1,y) + wy(x,y) + wy(x,y+1)
        
        st(A,-wy(x,y+1),i,i+m,x,y+1)
        st(A,-wy(x,y),  i,i-m,x,y-1)
        st(A,-wx(x+1,y),i,i+1,x+1,y)
        st(A,-wx(x,y),  i,i-1,x-1,y)
   
    end
    
    return A
end
```

```julia
function exactsolve(img,lx,ly)
    A = getTikhonovMatrix(img,lx,ly)
    b = vec(img)
    return A\b
end;

function seidelsolve(img,lx,ly)
    A = getTikhonovMatrix(img,lx,ly)
    b = vec(img)
    x = copy(b)
    gauss_seidel!(x,A,b)
    return x
end;

function getKernel(img,lx,ly)
    n,m = size(img)
    A = getTikhonovMatrix(img,lx,ly)
    mK = inv(A)
    K = zeros(n,m,n,m) #kernel K(x,y,u,v)
    for x=1:n,y=1:m,u=1:n,v=1:m
        K[x,y,u,v] = mK[(y-1)*n + x, (v-1)*n + u]
    end
    return K
end
```

# Extracting domain transform

Suppose $K(x,y,u,v) = a(x,y)b(u,v)h(t(x,y)-t(u,v))$

Here $h(0)=1$ (unit height kernel), also assume that

$h(p) = h(-p)$


Then, the kernel symmetry $K(u,v,x,y) = K(x,y,u,v)$ implies

$a(x,y)b(u,v) = a(u,v)b(x,y)$

so

$K(x,y,x,y) K(u,v,u,v) = a(x,y)b(u,v) a(u,v) b(x,y)$

this means 

$ K(x,y,u,v) K(u,v,x,y) = K(x,y,x,y) K(u,v,u,v) h(t(x,y)-t(u,v))^2$,

so

$K(x,y,u,v) = \sqrt{K(x,y,x,y) K(u,v,u,v)} h(t(x,y)-t(u,v))$

$K(x,y,u,v) = \sqrt{K(x,y,x,y) K(u,v,u,v)} h(dist((x,y),(u,v)))$

```julia
function unnormalize(K)
    res = copy(K)
    n,m = size(K)[1:2]
    
    for x = 1:n, y=1:m, u=1:n, v=1:m
        res[x,y,u,v] = K[x,y,u,v]/sqrt(K[x,y,x,y]*K[u,v,u,v])
    end
    
    return res
end;
```

```julia
uK = unnormalize(K);
```

# Transform

```julia
function exactTransform(K)
    N = size(K)[1]

    dc = zeros(N)
    da = zeros(N)
    
    for i =2:N
        dc[i] = log(K[N,i]) - log(K[N,i-1])
    end
   
    for i =N-1:-1:2
        da[i] = abs(log(K[1,i]) - log(K[1,i-1]))
    end
 
    return cumsum((dc .+ da) ./2)
end;
```

# Using general metrics

Assume the general form for $p,q \in \mathbb{R}^2$

$K(p,q) = a(p)a(q)h(dist(p,q))$

The function $\Omega(p,q):=\frac{K(p,q)}{a(p)a(q)}$ is known for the Tikhonov filter

This means we require an invertible function $h:\mathbb{R}^+ \rightarrow \mathbb{R}^+$ such that

$dist(p,q) = h^{-1}(\Omega(p,q))$ and this function, being a metric, must satisfy

a) $h^{-1}(\Omega(p,q)) = 0 \iff p=q$ (always true as long as $h^{-1}(1)=0$, unit height kernel and injective) - Identity of indiscernibles

b) $h^{-1}(\Omega(p,q)) = h^{-1}(\Omega(q,p))$ (always true from construction of $\Omega$) -Symmetry

c) $h^{-1}(\Omega(p,r)) \leq h^{-1}(\Omega(p,q)) + h^{-1}(\Omega(q,r)) $ - 	Triangle inequality

d*) $h^{-1}(\Omega(p,r)) = \min_{q \in V_r} h^{-1}(\Omega(p,q)) + h^{-1}(\Omega(q,r))$ - Geodesic distance (not required by metrics, but useful for practical use)

```julia
function exponentialTest(uK)
    n,m = size(uK)[1:2]
    res = 1
    
    for y3=1:m, x3=1:n, y2=1:m, x2=1:n, y1=1:m, x1=1:n
        res = min(res, log(uK[x1,y1,x3,y3]) - log(uK[x1,y1,x2,y2]) - log(uK[x2,y2,x3,y3]))
    end
    
    return res
end;

function gaussianTest(uK)
    n,m = size(uK)[1:2]
    res = 1
    for y3=1:m, x3=1:n, y2=1:m, x2=1:n, y1=1:m, x1=1:n
        res = min(res, -sqrt(abs(log(uK[x1,y1,x3,y3]))) + sqrt(abs(log(uK[x1,y1,x2,y2]))) + sqrt(abs(log(uK[x2,y2,x3,y3]))))
    end
    
    return res
end
```

```julia
function exponentialGeodesicPath(uK,p,r; alpha = 1, beta=1,threshold = 1)
    n,m = size(uK)[1:2]
    
    x1,y1 = p
    x3,y3 = r
    
    res = 1
    
    out = zeros(n,m)
    
    dist(x1,y1,x2,y2) = abs(log(uK[x1,y1,x2,y2])/alpha)^(1/beta)
    
    for y2 = 1:n, x2=1:m
        out[y2,x2] = dist(x1,y1,x2,y2) + dist(x2,y2,x3,y3) - dist(x1,y1,x3,y3)
    end
    
    out = min.(out,threshold)
    
    return out 
end;
```

```julia
py.imshow(exponentialGeodesicPath(uK,(10,30),(75,30), alpha=1 ,beta=0.92, threshold = 1), cmap = "jet")
py.colorbar()
```
